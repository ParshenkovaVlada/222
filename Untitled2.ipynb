```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных с преобразованием в числовой формат
url = "https://drive.google.com/uc?export=download&id=130uKatSQ7Tpm4cAED2Q5H5C8WQSdDm8R"
wind_speed = pd.read_excel(url, header=None).apply(pd.to_numeric, errors='coerce').values.flatten()

# Устранение ошибок в приобразовании
wind_speed = wind_speed[~np.isnan(wind_speed)]

def wind_speed_to_power(speed):
    if speed < 5 or speed > 25:
        return 0
    elif 5 <= speed <= 20:
        return 0.0017 * speed ** 3
    elif 20 < speed <= 25:
        return 0.0017 * 20 ** 3  # Максимальная мощность при 20 м/с

# Векторизованная версия функции
wind_speed_to_power_vec = np.vectorize(wind_speed_to_power)

# График зависимости выработки от скорости ветра
x = np.arange(0, 30, 0.1)
y = wind_speed_to_power_vec(x)

fig1, ax1 = plt.subplots()
ax1.plot(x, y)
ax1.grid(True)
ax1.set_xlabel('Скорость ветра, м/с')
ax1.set_ylabel('Мощность ВЭУ, МВт')
ax1.set_title('Зависимость мощности ВЭС от скорости ветра')
plt.show()

# Расчет мощности для каждого часа
power = wind_speed_to_power_vec(wind_speed)

# График среднесуточной выработки (почасовой)
hours = np.arange(len(wind_speed))
day = hours/24
fig2, ax2 = plt.subplots(figsize=(20, 10),)
ax2.plot(day, power, label='Мощность, МВт')
ax2.plot(day, wind_speed, label='Скорость ветра м/с')

ax2.grid(True)
ax2.set_xlabel('Время, дни')
ax2.set_ylabel('Скорость ветра/Мощность')
ax2.set_title('Поcуточная выработка ВЭС и скорость ветра')
ax2.legend()
plt.show()

# Расчет среднесуточной мощности
daily_avg_power = np.mean(power)
print(f"Среднесуточная выработка ВЭС: {daily_avg_power:.2f} МВт")
```
